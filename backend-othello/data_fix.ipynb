Import required libraries for data fix:

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

Basic data set cleaning:

```python
df = pd.read_csv(r"data/othello_training_data.csv")
```

```python
df = df.drop(columns=['game_id'])
```

Normalization of data set in order to make ML learning easier with normalized data that is more friendly for neural networks:

```python
scaler = MinMaxScaler()
```

```python
# normalizing only "score" column because otherwise it will be harder for neural network to learn if I also normalize whose turn is now
df['score'] = scaler.fit_transform(df[['score']])
```

Save file for later use in neural network:

```python
df.to_csv("data/othello_training_data_fixed.csv", index=False)
```
